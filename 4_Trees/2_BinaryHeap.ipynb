```python
class BinaryHeap():
    
    def __init__(self, heapType = 'max'):
        self.heap = [0]
        if heapType == ('max' or 'min'):
            self.heapType = heapType
        else:
            raise ValueError("heapType must be 'max' or 'min'.")
        self.size = 0
        
    def insert(k):
        self.heap.append(k)
        self.size += 1
        self.swapUp(self.size)
        
    def swapUp(self, node_ind):
        if node_ind == 1:
            pass
        
        node_val = self.heap[node_ind]
        parent_ind = self.size//2
        parent_val = self.heap[parent_ind]
        
        if self.heapType == 'max':
            if node_val>parent_val:
                self.heap[parent_ind] = node_val
                self.heap[node_ind] = parent_val
                self.swapUp(parent_ind)
        else:
            if node_val<parent_val:
                self.heap[parent_ind] = node_val
                self.heap[node_ind] = parent_val
                self.swapUp(parent_ind)
                
    def swapDown(selfm, node_ind):
        node_val = self.heap[node_ind]
        
        # Using a while loop instead of recursion this time.
        while node_ind*2 <= self.size:
            
            if self.heapType == 'max':
                mChild_ind = self.maxChild(node_ind)
                mChild_val = self.heap[mChild_ind]
                
                if node_val<mChild_val:
                    self.heap[mChild_ind] = node_val
                    self.heap[node_ind] = mChild_val
                    node_ind = mChild_ind
            
            elif self.heapType == 'min':
                mChild_ind = self.minChild(node_ind)
                mChild_val = self.heap[mChild_ind]
                
                if node_val>mChild_val:
                    self.heap[mChild_ind] = node_val
                    self.heap[node_ind] = mChild_val
                    node_ind = mChild_ind
    
    def maxChild(self, node_ind):
        child1 = node_ind*2
        child2 = node_ind*2+1
        
        if self.heap[child1] > self.heap[child2]:
            return child1
        else: 
            return child2
    
    def minChild(self, node_ind):
        child1 = node_ind*2
        child2 = node_ind*2+1
        
        if self.heap[child1] < self.heap[child2]:
            return child1
        else: 
            return child2
        
    def delTop(self):
        ret_val = self.heap[1]
        self.heap[1] = self.heap.pop()
        self.size -= 1
        self.swapDown()
        return ret_val
        
    def buildHeap(self, arr):
        self.heap = self.heap + arr
        self.size = len(arr)
        i = len(arr)//2
        while i > 0:
            self.swapDown(i)
            i -= 1
            
```
