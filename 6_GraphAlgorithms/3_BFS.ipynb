```python
graph = {'A': set(['B', 'C', 'G']),
         'B': set(['A', 'D', 'E']),
         'C': set(['A', 'F']),
         'D': set(['B']),
         'E': set(['B', 'F', 'I']),
         'F': set(['C', 'E', 'H']),
         'G': set(['A', 'F', 'H']),
         'H': set(['I']),
         'I': set(['B', 'F'])
        }
```

```python
def BFS(graph, start, end):
    
    current = start
    path = [start]
    queue = [(current, path)]
    
    while queue != []:
        
        print()
        print(queue)
        
        (current, path) = queue.pop(0)
        not_visited = graph[current] - set(path)
        
        print('Choose start-of-queue path: ', path)
        print(f'New nodes connected to {current}: ', not_visited)
        
        for nxt in not_visited:
        
            if nxt == end:
                yield path + [nxt]
            else:
                queue.append((nxt, path + [nxt]))
```

```python
list(BFS(graph, "A", "I"))
```
